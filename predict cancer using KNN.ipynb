# Converting Categorical Data

```python
df['diagnosis'].unique()
```

```python
diagnosis=pd.get_dummies(df['diagnosis'],drop_first=True)
diagnosis
```

```python
df.head()
```

```python
df.drop(['diagnosis'],axis=1,inplace=True)
```

```python
df=pd.concat([df,diagnosis],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['concave points_worst','perimeter_worst','radius_worst','concave points_mean','perimeter_mean']]
Y=df['M']
```

```python
X
```

# Data Normalisation

This brings all of the data to the same scale so it is easier for the algorithm to compute it

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.33)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Predicting using K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Starting off with a random number of neighbours**

```python
testknn=KNeighborsClassifier(n_neighbors=5)
```

```python
testknn.fit(X_train, Y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(testknn,X,Y,cv=5,scoring='accuracy')
```

**Checking algorithm with many different K values to determine the best one**

```python
df.shape
```

```python
569/5
```

```python
113.8*4
```

```python
ks=list(range(1,455,2))
```

```python
cvScores=[]
```

```python
for k in ks:
    knn=KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(knn,X,Y,cv=5,scoring='accuracy').mean()
    cvScores.append(score)
```

```python
cvScores
```

```python
max(cvScores)
```

```python
cvScores.index(max(cvScores))
```

**Finalising the algorithm with K=21**

```python
finalKNN= KNeighborsClassifier(n_neighbors=21)
```

```python
X = scaler.transform(X)
```

```python
finalKNN.fit(X,Y)
```

```python
X
```

# Saving the Algorithm & Scaler

```python
import pickle
pickle.dump(finalKNN,open('KNN Model.pkl','wb'))
pickle.dump(scaler, open('Scaler.pkl', 'wb'))
```
